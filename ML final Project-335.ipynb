```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

X = df.drop(columns=['CustomerID', 'Conversion','Age','Income','AdvertisingTool','AdvertisingPlatform'])
y = df['Conversion']

X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
print('\nAccuracy Score:')
print(accuracy_score(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to features and labels
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check new class distribution
from collections import Counter
print("Resampled class distribution:", Counter(y_resampled))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Train-test split on resampled data
X_train_sm, X_test_sm, y_train_sm, y_test_sm = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42
)

# Train the model
model = LogisticRegression(solver='liblinear')
model.fit(X_train_sm, y_train_sm)

# Predictions
y_pred_sm = model.predict(X_test_sm)

# Evaluate
print(classification_report(y_test_sm, y_pred_sm))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

dtc = DecisionTreeClassifier(max_depth=10, min_samples_split=5, min_samples_leaf=2, criterion='entropy', random_state=50)
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1, random_state=36)
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=58) 
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)
rf_clf_acc = accuracy_score(y_test, y_pred_rf)
rf_clf_acc
```

```python
import gradio as gr
import pandas as pd
import joblib

model = joblib.load("logreg_smoteenn_model.pkl")
scaler = joblib.load("scaler.pkl")
feature_names = joblib.load("feature_names.pkl")

def predict_conversion(
    Gender, CampaignChannel, CampaignType, AdSpend, ClickThroughRate, ConversionRate,
    WebsiteVisits, PagesPerVisit, TimeOnSite, SocialShares, EmailOpens, EmailClicks,
    PreviousPurchases, LoyaltyPoints
):
    input_data = {
        'AdSpend': [AdSpend],
        'ClickThroughRate': [ClickThroughRate],
        'ConversionRate': [ConversionRate],
        'WebsiteVisits': [WebsiteVisits],
        'PagesPerVisit': [PagesPerVisit],
        'TimeOnSite': [TimeOnSite],
        'SocialShares': [SocialShares],
        'EmailOpens': [EmailOpens],
        'EmailClicks': [EmailClicks],
        'PreviousPurchases': [PreviousPurchases],
        'LoyaltyPoints': [LoyaltyPoints],
        'Gender_Male': [1 if Gender == 'Male' else 0],
        'CampaignChannel_PPC': [1 if CampaignChannel == 'PPC' else 0],
        'CampaignChannel_Social Media': [1 if CampaignChannel == 'Social Media' else 0],
        'CampaignType_Retention': [1 if CampaignType == 'Retention' else 0],
        'CampaignType_Conversion': [1 if CampaignType == 'Conversion' else 0]
    }

    input_df = pd.DataFrame(input_data)
    for col in feature_names:
        if col not in input_df.columns:
            input_df[col] = 0
    input_df = input_df[feature_names]

    scaled_input = scaler.transform(input_df)
    prediction = model.predict(scaled_input)[0]

    if prediction == 1:
        return "<div style='background-color:#e6ffe6; padding:15px; width: 200px;'>✅ Likely to convert.</div>"
    else:
        return "<div style='background-color:#ffe6e6; padding:15px; width: 200px;'>❌ Unlikely to convert.</div>"

with gr.Blocks(title="📊 Conversion Predictor App") as demo:
    gr.Markdown("## 🎯 Digital Marketing Campaign Optimizer")
    with gr.Row():
        with gr.Column():
            gender = gr.Dropdown(['Male', 'Female'], label="Gender")
            channel = gr.Dropdown(['Email', 'PPC', 'Social Media'], label="Campaign Channel")
            ctype = gr.Dropdown(['Awareness', 'Retention', 'Conversion'], label="Campaign Type")
        with gr.Column():
            adspend = gr.Slider(100, 10000, step=100, label="Ad Spend ($)")
            ctr = gr.Slider(0.01, 0.3, step=0.01, label="Click Through Rate")
            conv_rate = gr.Slider(0.01, 0.2, step=0.01, label="Conversion Rate")
        with gr.Column():
            visits = gr.Slider(0, 50, step=1, label="Website Visits")
            pages = gr.Slider(1.0, 10.0, step=0.1, label="Pages Per Visit")
            time = gr.Slider(0.5, 15.0, step=0.1, label="Time on Site (mins)")
        with gr.Column():
            shares = gr.Slider(0, 100, step=1, label="Social Shares")
            opens = gr.Slider(0, 20, step=1, label="Email Opens")
            clicks = gr.Slider(0, 10, step=1, label="Email Clicks")
        with gr.Column():
            purchases = gr.Slider(0, 10, step=1, label="Previous Purchases")
            points = gr.Slider(0, 5000, step=100, label="Loyalty Points")
            predict_btn = gr.Button("🔍 Predict")

        result = gr.HTML(label="Prediction Result")

    predict_btn.click(
        fn=predict_conversion,
        inputs=[
            gender, channel, ctype, adspend, ctr, conv_rate,
            visits, pages, time, shares, opens, clicks, purchases, points
        ],
        outputs=result
    )

demo.launch()
```
